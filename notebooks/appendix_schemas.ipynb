```python
import hopsworks
project = hopsworks.login(
    host="eu-west.cloud.hopsworks.ai",
    project="London_traffic"
)
fs = project.get_feature_store()
feature_groups = [
    ("traffic_flow_fg", 1),
    ("traffic_points_metadata", 1),
    ("labels_speed_ratio_fg", 1),
    ("traffic_temporal_fg", 2),
    ("weather_10m_fg", 1),
    ("tfl_disruptions_10m_fg", 1),
]

for name, version in feature_groups:
    fg = fs.get_feature_group(name=name, version=version)
    
    print("=" * 90)
    print(f"Feature Group: {fg.name} | Version: {fg.version}")
    print(f"Description: {fg.description}")
    print("- Schema -")
    
    for f in fg.features:
        print(f"{f.name}: {f.type}")

```

```
from pathlib import Path
import pandas as pd

obs_path = Path('data/processed/observations.parquet')
assert obs_path.exists(), 'Run build_dataset first.'

model_15 = Path('models/hgb_h15.joblib')
model_30 = Path('models/hgb_h30.joblib')
print('Models exist:', model_15.exists(), model_30.exists())
```

```
from traffic_forecast.inference import predict_for_point

df = pd.read_parquet(obs_path)
one_point = df['point_id'].astype(str).value_counts().index[0]

pred15 = predict_for_point(model_15, observations_path=obs_path, point_id=one_point)
pred30 = predict_for_point(model_30, observations_path=obs_path, point_id=one_point)
pred15, pred30
```
